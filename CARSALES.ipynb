```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Car Sales.csv")
```

```python
print(data)
```

ABOUT DATASET:
The dataset consists of several columns:

**week_id:** Represents the week number or identifier.

**date:** Represents the date of the week.

**sessions:** Indicates the number of sessions or visits to the website for that particular week.

**online_leads_actuals:** Represents the number of online leads or inquiries generated during that week.

**test_drive_request:** Indicates the number of test drive requests made by customers.

**offer_request:** Represents the number of offer requests received during that week. It seems like this column might contain categorical data indicating the level of the offer request (e.g., low, medium, high).

**total_configs_finished:** Indicates the total number of configurations finished, which could represent the number of customizations made by customers.

**keep_me_updated:** This column seems to indicate whether customers opted to stay updated (binary: 0 or 1).

**dealer_locator:** Represents the usage of the dealer locator service.

# VISUALIZATION

**1. Creating a line chart to visualize the trend of total configurations finished over the different weeks.**

```python
# Group data by week_id and calculate sum of total_configs_finished
grouped_data = data.groupby('week_id')['total_configs_finished'].sum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data.values, marker='o', color='blue')
plt.title('Total Configurations Finished by Week')
plt.xlabel('Week ID')
plt.ylabel('Total Configurations Finished')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**2. Creating a pie chart representing the number of sessions per week**

```python
# Group data by week_id and calculate sum of sessions
sessions_per_week = data.groupby('week_id')['sessions'].sum()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(sessions_per_week.values, labels=sessions_per_week.index, autopct='%1.1f%%', startangle=140)
plt.title('Number of Sessions per Week')
plt.axis('equal')
plt.show()
```

**3. Creating a barchart to visualize the distribution of offer requests per week**

```python
# Group data by week_id and calculate sum of offer_request
offer_requests_per_week = data.groupby('week_id')['offer_request'].sum()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(offer_requests_per_week.index, offer_requests_per_week.values, color='skyblue')
plt.title('Offer Requests per Week')
plt.xlabel('Week')
plt.ylabel('Offer Requests')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**4. Finding the top 5 dates with the highest number of online leads,**
